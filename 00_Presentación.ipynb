#### Software necesario
* `R`: https://www.r-project.org/ Es un lenguaje para cómputo estadístico, de licencia gratuita, y con una comunidad de desarrolladores amplia y activa.
* `JAGS`: http://mcmc-jags.sourceforge.net/ **J**ust **A**nother **G**ibbs **S**ampler es un programa diseñado para implementar diferentes modelos estadísticos desde la perspectiva Bayesiana y encontrar su solución. Está escrito en `C++` y es independiente de `R`, aunque ambos pueden vincularse con la librería siguiente.
* `R2jags`: librería de `R` que permite utilizar `JAGS` en el mismo ambiente de trabajo. Para instalarla es necesario 1. instalar `JAGS` en la computadora, y 2. ejecutar `install.packages('R2jags')` en la consola de `R`.

#### Software recomendado 
* `RStudio`: https://rstudio.com/ IDE para `R` que permite trabajar con dicho lenguaje cómodamente. 
* `Jupyter/Jupyter Notebooks`: https://jupyter.org/ Proyecto escrito en `Python` diseñado para generar documentos (notebooks) que incluyen prosa y código (este documento está escrito en dicho formato: **.ipynb**). No sólo permite incluir código en `R`: también pueden generar cuadernillos con código en `Python`, `Julia`, `C`, entre muchos otros lenguajes actuales.
    * Para instalar `Jupyter` primero (o al mismo tiempo) tienen que instalar `Python` (les recomiendo que se aseguren de instalar la versión `3` o mayor, e.g. `Python 3.8`). Existen muchas alternativas para hacerlo y cuál es la mejor depende de su sistema operativo: 
        * `Linux` y `Mac` generalmente ya tienen una distribución de `Python` actualizada, lo cual sólo vuelve necesario instalar `Jupyter` como librería adicional. En dichos sistemas operativos esta instalación puede completarse ejecutando `pip3 install --user jupyter` en la terminal del sistema (aunque tampoco es la única opción: también pueden instalar `Anaconda` como se sugiere a continuación). 
        * `Windows` generalmente no incluye `Python`. Pueden buscar "distribuciones de Python", instalar alguna, y después buscar cómo se instalan librerías adicionales en dicha distribución. O pueden instalar `Anaconda` https://www.anaconda.com/products/individual, que es una distribución de `Python` que ya incluye `Jupyter` (entre otras librerías enfocadas a ciencia de datos). Recomiendo esta última porque _parece_ ser la más cómoda, aunque personalmente no he seguido sus pasos desde hace varios años.
        * En general, independientemente de cuál sistema utilizan, les recomiendo buscar tutoriales de instalación pertinentes **antes** de descargar y ejecutar cualquier cosa. Hay muchas páginas y videos con instrucciones detalladas para prácticamente cualquier sistema operativo contemporáneo.
    * Después de instalar `Jupyter` es necesario vincularlo con `R`. Esto se hace instalando varias librerías adicionales en `R`. Cuáles necesitan depende de cuáles ya tienen y cuáles toca actualizar, aunque generalmente estas instrucciones (en la consola de `R`) son suficientes:
    
        `install.packages(c('repr', 'IRdisplay', 'evaluate', 'crayon', 'pbdZMQ', 'devtools', 'uuid', 'digest'))`
        
        `devtools::install_github('IRkernel/IRkernel')`
        
        `IRkernel::installspec()`
    
        si la instalación fue correcta, deberían ver la opción "`R`" al generar un cuadernillo nuevo. 
        
        (P.S. ¡También hay muchos tutoriales para vincular `R` con `Jupyter`!)
