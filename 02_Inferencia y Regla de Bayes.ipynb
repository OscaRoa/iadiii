# Inferencia

### Construyendo distribuciones conjuntas

En el cuadernillo anterior presentamos la definición de incertidumbre condicional:

$$P(A|B)=\frac{P(A,B)}{P(B)},$$

en donde la marginal $P(B)=\sum_{a}P(A,B)$ puede calcularse a partir de la misma función conjunta $P(A,B)$. En otras palabras, si conocemos la conjunta conocemos las marginales y podemos calcular cualquier condicional.

En aquellos ejemplos partimos de distribuciones conjuntas previamente especificadas y "extrajimos" marginales y condicionales. En la práctica, sin embargo, no siempre resulta práctico intentar especificar _directamente_ la distribución conjunta, aunque necesitamos dicha distribución para entender la relación entre las variables involucradas. Por ejemplo, si está en nuestro interés estudiar la relación entre _conducir un automóvil después de una fiesta_ y _tener un accidente_, variables con los niveles:

$$A=\begin{cases}
A=0:  \textrm{no tener accidente}\\
A=1:  \textrm{tener accidente}\\
\end{cases}$$

$$B=\begin{cases}
B=1:  \textrm{manejar sin haber consumido alcohol}\\
B=2:  \textrm{manejar con el máximo de alcohol permitido}\\
B=3:  \textrm{manejar con más del máximo de alcohol permitido}\\ 
\end{cases}$$

sabemos que la conjunta estará definida en las siguientes dimensiones:

|P(A,B) |A=0|A=1|
|:-----:|:-:|:-:|
|**B=1**|?  |?  |
|**B=2**|?  |?  |
|**B=3**|?  |?  |

Asignar probabilidades a cada una de estas seis entradas, sin embargo, puede resultar difícil y poco intuitivo pues tendríamos que responder preguntas del estilo: _¿qué porcentaje de la población maneja sin alcohol **y** tiene una accidente?_, o _¿qué proporción de conductores manejan con el máximo de alcohol permitido **y** no tienen un accidente?_, etcétera.

Desde luego podríamos diseñar un estudio enfocado a medir dichas proporciones directamente, pero también es posible utilizar otras piezas de información (que tal vez ya estén disponibles) para "construirla". Por ejemplo, **1.** podríamos consultar a los _valet parking_ en las zonas de bares: como devuelven los coches y ven a los conductores, seguramente tienen una idea clara de la incertidumbre marginal $P(B)$, aunque no tengan información sobre la suerte de los conductores una vez que salen del estacionamiento. Por otra parte, **2.** quizá alguna dependencia del gobierno o algún equipo de investigación ha estudiado la incidencia de accidentes en cada tipo de conductor, digamos tomando a 100 conductores sobrios, 100 con el máximo de alcohol permitido, y 100 con más del máximo, y registrando cuántos en cada grupo tienen un accidente la noche de la medición. Esta información permitiría aproximar la distribución condicional $P(A|B)$, aunque no diría nada sobre ninguna marginal.

Ambas distribuciones permiten calcular la conjunta $P(A,B)$ aprovechando que en la identidad

$$P(A|B)=\frac{P(A,B)}{P(B)},$$

podemos despejar $P(A,B)$ multiplicando ambos lados de la ecuación por $P(B)$:

$$P(A|B)P(B)=P(A,B)$$

Por ejemplo, suponiendo que los valet parking nos revelan que

$$P(B)=\begin{cases}
P(B=1)=0.2\\
P(B=2)=0.5\\
P(B=3)=0.3,\\ 
\end{cases}$$

y que el equipo de investigación tiene datos que sugieren

$$P(A|B)=\begin{cases}
P(A|B=1)=\begin{cases}
    P(A=0|B=1)=0.9\\
    P(A=1|B=1)=0.1\\
    \end{cases}\\\\
P(A|B=2)=\begin{cases}
    P(A=0|B=2)=0.6\\
    P(A=1|B=2)=0.4\\
    \end{cases}\\\\
P(A|B=3)=\begin{cases}
    P(A=0|B=3)=0.2\\
    P(A=1|B=3)=0.8,\\
    \end{cases}
\end{cases}$$

algunas entradas de la conjunta son

$$P(A,B)=\begin{cases}
P(A=0,B=1)=P(A=0|B=1)P(B=1)=(0.9)(0.2)=0.18\\
P(A=1,B=1)=P(A=1|B=1)P(B=1)=(0.1)(0.2)=0.02\\
    \vdots\\
P(A=1,B=3)=P(A=1|B=3)P(B=3)=(0.8)(0.3)=0.24\\ 
\end{cases}$$

y la conjunta completa:

|P(B)|||P(A=0\|B)|P(A=1\|B)|||P(A,B) |A=0|A=1|
|:--:|:--:|:--:|:--:|:--:|:--:|:--:|:-----:|:-:|:-:|
|**0.2**|||**0.9**|**0.1**|||**B=1**|0.18  |0.02  |
|**0.5**|||**0.6**|**0.4**|||**B=2**|0.30  |0.20  |
|**0.3**|||**0.2**|**0.8**|||**B=3**|0.06  |0.24  |

Pueden comprobar que la función $P(A,B)$ resultante cumple

$$\sum_{a}\sum_{b}P(A,B)=1$$

Nuevamente, una vez que tenemos la conjunta podemos *a.* reproducir la marginal sobre $B$, *b.* calcular la marginal sobre $A$ (desconocida hasta este momento en nuestro estudio), y *c.* calcular cualquier condicional, en cualquier dirección.

La ruta que seguimos para construir la conjunta en el ejemplo anterior no es la única alternativa. También podríamos, por ejemplo, viajar a otra ciudad en donde no existen los _valet_, restricción que nos impide tener acceso a $P(B)$, pero en la que el control de tráfico gubernamental es altamente sofisticado y mediante rastreo satelital permite responder, en cierta noche, _de cada 100 coches circulando, ¿cuántos tienen accidente y cuántos no?_ Es decir, en este caso tenemos acceso a $P(A)$, aunque con esta primera medición no sabemos nada sobre el estado de alcohol de los conductores. Aparte, podríamos visitar los centros de detención y los hospitales de la ciudad para registrar el nivel de alcohol en la sangre de aquellos conductores que tuvieron un accidente, aproximando así $P(B|A=1)$ y, aprovechando la información satelital, podríamos visitar a los conductores que llegaron a salvo a su casa y obligarlos a tomar la prueba de alcoholímetro para estimar $P(B|A=0)$.

En otras palabras, en este segundo ejemplo, a diferencia del anterior, calcularemos la conjunta a partir de la marginal sobre $A$, aprovechando que en la definición

$$P(B|A)=\frac{P(A,B)}{P(A)}$$

nuevamente podemos despejar $P(A,B)$, esta vez multiplicando ambos lados por $P(A)$:

$$P(B|A)P(A)=P(A,B),$$

identidad conocida como **regla del producto**.
        
Suponiendo que en aquella ciudad los satélites registran:

$$P(A)=\begin{cases}
    P(A=0)=\frac{54}{100}\\\\
    P(A=1)=\frac{46}{100}
\end{cases}$$

y que al visitar a los 9 conductores que llegaron a sus casas sin accidentes, y al visitar a los 23 accidentados en hospitales, descubrimos, respectivamente:

$$P(B|A)=\begin{cases}
    P(B|A=0)=\begin{cases}
        P(B=1|A=0)=\frac{3}{9}\\\\
        P(B=2|A=0)=\frac{5}{9}\\\\
        P(B=3|A=0)=\frac{1}{9}
        \end{cases}\\\\
    P(B|A=1)=\begin{cases}
        P(B=1|A=1)=\frac{1}{23}\\\\
        P(B=2|A=1)=\frac{10}{23}\\\\
        P(B=3|A=1)=\frac{12}{23}
    \end{cases}
\end{cases}$$

Podemos calcular las primeras entradas de la conjunta como:

$$P(A,B)=\begin{cases}
P(A=0,B=1)=P(B=1|A=0)P(A=0)=\left(\frac{3}{9}\right)\left(\frac{54}{100}\right)=\frac{162}{900}=0.18\\\\
P(A=1,B=1)=P(B=1|A=1)P(A=1)=\left(\frac{1}{23}\right)\left(\frac{46}{100}\right)=\frac{46}{2300}=0.02\\
    \vdots\\
P(A=1,B=3)=P(B=3|A=1)P(A=1)=\left(\frac{12}{23}\right)\left(\frac{46}{100}\right)=\frac{552}{2300}=0.24 
\end{cases}$$

Y la conjunta completa:

||A=0|A=1|
|:-----:|:-:|:-:|
|**P(A)**|**54/100**|**46/100**|
|| | |
|**P(B=1\|A)**|**3/9** | **1/23**|
|**P(B=2\|A)**|**5/9** | **10/23**|
|**P(B=3\|A)**|**1/9** | **12/23**|
|| | |
|**P(A,B)**|**A=0**|**A=1**|
| | |
|**B=1**|0.18  |0.02  |
|**B=2**|0.30  |0.20  |
|**B=3**|0.06  |0.24  |

Resultado que reproduce la conjunta de la primera ciudad. Desde luego, no hay razones para que dicha distribución sea igual en dos ciudades reales, pero hemos elegido estos ejemplos para enfatizar que una sola incertidumbre conjunta puede construirse de dos maneras diferentes:

$$P(A|B)P(B)=P(A,B)$$

y

$$P(B|A)P(A)=P(A,B)$$

### Regla de Bayes

En los dos ejemplos anteriores estudiamos los dos caminos posibles para construir una conjunta de dos dimensiones. Ambas rutas expresan la distribución conjunta $P(A,B)$ como el producto de **una marginal** por la **condicional sobre la otra variable**:

$$P(B|A)P(A)=P(A,B)=P(A|B)P(B),$$

lo cual sugiere que existe cierta relación **específica** entre las condicionales de cada variable, **idea** conocida como _**Regla de Bayes**_. En concreto, dividiendo ambos lados de la ecuación entre $P(A)$:

$$P(B|A)=\frac{P(A|B)P(B)}{P(A)}$$

y recordando que la marginal $P(A)$ puede calcularse a partir de la conjunta como

$$P(A)=\sum_{b}P(A,B),$$

en donde nuevamente la conjunta puede expresarse como $P(A,B)=P(A|B)P(B)$, lo cual resulta en la expresión para la marginal:

$$P(A)=\sum_{b}P(A|B)P(B)$$

Sustituyendo en la Regla de Bayes:


$$P(B|A)=\frac{P(A|B)P(B)}{\sum_{b}P(A|B)P(B)}$$

Esta expresión permite apreciar la idea completa de la Regla: la incertidumbre actualizada (o _posterior_) sobre alguna de las variables $P(B|A)$, resultado de observar la otra, puede calcularse a partir de **a.** la incertidumbre inicial (o marginal, o _a priori_) sobre dicha variable $P(B)$ y de **b.** la incertidumbre condicional sobre la otra variable dada la variable de interés $P(A|B)$.   

#### Ejemplo: Prueba de COVID

En un escenario **ficticio**, imaginen una prueba diseñada para detectar la presencia o ausencia del virus. Esta prueba tiene una tasa de "**positivos correctos**" de **80%**, lo cual quiere decir que de cada 100 personas _que realmente_ tienen el virus la prueba responde "positivo" en 80 de ellas. Además, también tiene una tasa de "**falsos positivos**" de **10%**, lo cual indica que de cada 100 personas que _no tienen_ el virus el test responde "positivo" para 10 de ellas. Finalmente, aparte de la información sobre el desempeño de la prueba, sabemos que la incidencia del virus es relativamente baja en la población, en tanto que aproximadamente **1 de cada 100** personas tienen el virus.

Imaginen que se aplican la prueba y ésta dice "**positivo**": ¿qué tan probable es que tengan el virus?

Precisando las variables aleatorias del problema:

$$\textrm{Resultado del test: }T\in\{"+","-"\}$$

$$\textrm{Presencia o ausencia del virus: }V\in\{\textrm{"sí"},\textrm{"no"}\}$$

Como hemos observado la variable $T="+"$ el problema de inferencia consiste en especificar la incertidumbre condicional sobre la variable $V$ a partir de dicha observación. Es decir, buscamos calcular $P(V|T="+")$. Utilizando la Regla de Bayes:

$$P(V|T)=\frac{P(T|V)P(V)}{\sum_{v}P(T|V)P(V)}$$

Calculando la probabilidad marginal $P(T="+")$:

$$\begin{align}
P(V|T="+")&=\frac{P(T="+"|V)P(V)}{\sum_{v}P(T="+"|V)P(V)}\\\\
          &=\frac{P(T="+"|V)P(V)}{P(T="+"|V=\textrm{sí})P(V=\textrm{sí})+P(T="+"|V=\textrm{no})P(V=\textrm{no})}\\\\
          &=\frac{P(T="+"|V)P(V)}{(0.80)(0.01)+(0.10)(0.99)}\\\\
          &=\frac{P(T="+"|V)P(V)}{(0.008)+(0.099)}\\\\
          &=\frac{P(T="+"|V)P(V)}{0.107}
\end{align}$$

Y calculando la incertidumbre actualizada sobre $V$:

$$\begin{align}
P(V=\textrm{sí}|T="+")&=\frac{P(T="+"|V=\textrm{sí})P(V=\textrm{sí})}{0.107}\\\\
                      &=\frac{(0.80)(0.01)}{0.107}\\\\
                      &\approx0.075
\end{align}$$

Es decir, después de observar un resultado positivo de la prueba deberíamos concluir que la probabilidad de tener el virus es aproximadamente **7.5%**. Desde luego el otro escenario posible en $V$ contiene el complemento restante:

$$\begin{align}
P(V=\textrm{no}|T="+")&=\frac{P(T="+"|V=\textrm{no})P(V=\textrm{no})}{0.107}\\\\
                      &=\frac{(0.10)(0.99)}{0.107}\\\\
                      &\approx0.925,
\end{align}$$

o bien, después de un resultado positivo tenemos **92.5%** de probabilidad de **no** tener el virus.

Este resultado contraintuitivo resalta la importancia de trabajar con herramientas que permitan traducir y manipular la incertidumbre con reglas estrictas y precisas, en tanto que muchas veces nuestra "intuición probabilística" puede apuntar hacia conclusiones incorrectas. La explicación intuitiva de la probabilidad posterior tan baja para virus es que, como el virus es "raro", o bien es poco probable estar infectado desde antes de tomar la prueba, incluso aunque tengamos evidencia "fuerte" en favor del virus la conclusión actualizada sobre virus debe tomar en cuenta la baja probabilidad inicial. Sin embargo, esto no quiere decir que la prueba sea completamente inútil: noten que después de tener un resultado positivo la probabilidad de tener el virus cambia de $0.01$ a $0.07$, ¡un incremento sustancial! La Regla de Bayes, o bien Teoría de Probabilidad, especifican _exactamente_ cómo mezclar ambas piezas de conocimiento (virus raro y prueba eficiente) para construir una conclusión "racional".

Resumiendo, la Regla de Bayes permite calcular la distribución $P(V|T)$ a partir de la (única) región de la conjunta necesaria para hacerlo. En el ejemplo del resultado positivo:

||P(V)||P(T=+\|V)|P(T=-\|V)|P(V,T)|T=+|T=-||P(V\|T=+)|P(V\|T=-)|
|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|
|**V=sí**|0.01||0.80|?|**V=sí**|0.008|?||0.075|?|
|**V=no**|0.99||0.10|?|**V=no**|0.099|?||0.925|?|
|||||||**0.107**|||||

(Aunque, como estudiaremos a lo largo del curso, siempre es recomendable examinar toda la conjunta en la medida de lo posible, porque representa la relación fundamental entre las variables de interés en cualquier modelo.)

#### Ejemplo: Moneda Justa y Tramposa

Consideren dos monedas, una normal y otra que tiene dos águilas. Las ponemos en una bolsa oscura, revolvemos, y tomamos una al azar y, sin examinar sus dos caras, echamos un volado con ella. Observamos "águila". ¿Qué moneda sacamos de la bolsa? O más preciso, _¿qué tan probable es haber sacado cada moneda?_

Variables aleatorias:

$$\mathrm{Moneda\ (\textbf{no observable}):\ }M\in\{\mathrm{justa},\mathrm{tramposa}\}$$

$$\mathrm{Volado\ (\textbf{observable}):\ }V\in\{\mathrm{sol},\mathrm{águila}\}$$

Después de observar $V=\mathrm{águila}$ el problema de inferencia consiste en especificar la incertidumbre actualizada sobre $M$, la variable no observable:

$$P(M|V)=\frac{P(V|M)P(M)}{\sum_{m}P(V|M)P(M)}$$

Calculando la probabilidad marginal $P(V=\mathrm{águila})$:

$$\begin{align}
P(M|V)&=\frac{P(V|M)P(M)}{\sum_{m}P(V=\mathrm{águila}|M)P(M)}\\\\
          &=\frac{P(V|M)P(M)}{P(V=\mathrm{águila}|M=\mathrm{justa})P(M=\mathrm{justa})+P(V=\mathrm{águila}|M=\mathrm{tramposa})P(M=\mathrm{tramposa})}\\\\
          &=\frac{P(V|M)P(M)}{\frac{1}{2}\frac{1}{2}+\frac{1}{1}\frac{1}{2}}\\\\
          &=\frac{P(V|M)P(M)}{\frac{1}{4}+\frac{1}{2}}\\\\
          &=\frac{P(V|M)P(M)}{\frac{3}{4}}
\end{align}$$

Es decir, en este experimento la probabilidad de observar $\mathrm{águila}$ antes de sacar la moneda y echarla al aire es igual a $0.75$. Calculando la incertidumbre posterior sobre $M$:

$$\begin{align}
P(M=\mathrm{justa}|V=\mathrm{águila})&=\frac{P(V=\mathrm{águila}|M=\mathrm{justa})P(M=\mathrm{justa})}{\frac{3}{4}}\\
    &=\frac{\frac{1}{4}}{\frac{3}{4}}\\\\
    &=\frac{1}{3}\\\\
P(M=\mathrm{tramposa}|V=\mathrm{águila})&=\frac{P(V=\mathrm{águila}|M=\mathrm{tramposa})P(M=\mathrm{tramposa})}{\frac{3}{4}}\\
    &=\frac{\frac{1}{2}}{\frac{3}{4}}\\\\
    &=\frac{2}{3}
\end{align}$$

#### Ejemplo: Ladrón y ADN

Imaginen que un ladrón roba una joyería a mitad de la noche. La policía detiene a tres sospechosos, pero sabe que sólo uno es responsable porque las cámaras de vigilancia graban a una persona. Lo sospechosos son:

$$\mathrm{Ladrón:\ } L\in\ \begin{cases} 
L=1: \textrm{Pablo, que cuenta con antecedentes penales}\\
L=2: \textrm{Pedro, que tiene deudas importantes}\\
L=3: \textrm{Poncho, que está desempleado, y es hermano de Pedro}
    \end{cases} $$
    
A falta de mayor evidencia, la incertidumbre inicial de la policía con esta información es:

$$P(L)=\ \begin{cases} 
P(L=1)=0.70\\
P(L=2)=0.20\\
P(L=3)=0.10\\
    \end{cases} $$
    
Unos días después, sin embargo, el equipo de peritos encuentra un cabello en la escena del crimen y al examinar su ADN los especialistas reportan "evidencia del ADN de Poncho". Este reporte no es definitivo porque las técnicas y equipo del laboratorio no permiten llegar a una conclusión segura, y además es posible explicar la presencia de un cabello de Poncho sin suponerlo en la escena del crimen. Para cuantificar estos escenarios de manera precisa la policía utiliza la siguiente **_función de verosimilitud_**, como también se le conoce a la condicional $P(\textrm{Evidencia}|\textrm{Hipótesis})$:

$$P(E|L)=\ \begin{cases} 
P(E|L=1)=0.05\\
P(E|L=2)=0.60\\
P(E|L=3)=0.80\\
\end{cases} $$

Actualizando la incertidumbre sobre $L$:

$$\begin{align}
P(L|E)&=\frac{P(E|L)P(L)}{\sum_l P(E|L)P(L)}\\
      &=\frac{P(E|L)P(L)}{P(E|L=1)P(L=1)+P(E|L=2)P(L=2)+P(E|L=3)P(L=3)}\\\\
      &=\frac{P(E|L)P(L)}{(0.05)(0.70)+(0.60)(0.20)+P(0.80)P(0.10)}\\\\
      &=\frac{P(E|L)P(L)}{0.235}\\\\
P(L=1|E)&=\frac{P(E|L=1)P(L=1)}{0.235}\\\\
      &=\frac{0.035}{0.235}\\\\
      &\approx 0.149\\\\
P(L=2|E)&=\frac{P(E|L=2)P(L=2)}{0.235}\\\\
      &=\frac{0.120}{0.235}\\\\
      &\approx 0.511\\\\
P(L=3|E)&=\frac{P(E|L=3)P(L=3)}{0.235}\\\\
      &=\frac{0.080}{0.235}\\\\
      &\approx 0.340
\end{align}$$

Es decir, después de observar la evidencia del cabello el principal sospechoso ahora es $L=2: \textrm{Pedro}$, ¡aunque no era ni el principal sospechoso al inicio, ni el sospechoso bajo el cual la evidencia era más probable!

Noten que la **distribución a priori** $P(L)$, la **distribución posterior** $P(L|E)$, y la **función de verosimilitud** $P(E|L)$ están definidas sobre $L$. La primera representa el conocimiento sobre $L$ _antes_ de observar la evidencia $E$; la segunda representa el conocimiento actualizado sobre $L$ _después_ de tomar en cuenta la evidencia; y la tercera representa _qué tan plausible es la evidencia_ $E$ bajo cada posible escenario en el espacio de hipótesis $L$. Como hemos estudiado en este apunte, la Regla de Bayes, y en general las reglas y definiciones en Teoría de Probabilidad, especifican la relación entre estas funciones. En concreto: la distribución posterior es una "mezcla" entre la incertidumbre a priori y la función de verosimilitud: 

```r
L <- c(1,2,3) # Hypothesis space
P_L <- c(0.70,0.20,0.10) # Prior
P_E_d_L <- c(0.05,0.60,0.80) # Likelihood
P_L_d_E <- P_L*P_E_d_L/sum(P_L*P_E_d_L) # Posterior (by Bayes)
col_prior <- '#4d98c5'
col_post <- '#35618f'
col_like <- '#f5943b'
par(cex.axis=1.5)
plot(NULL,xlim=c(.5,3.5),ylim=c(0,1.2),ann=F,axes=F)
axis(1,at=L);axis(2,at=seq(0,1,.2),pos=0.5,las=1)
mtext('hipótesis (L)',1,line=3,cex=2)
mtext('probabilidad',2,line=2.5,cex=2,adj=.4)
points(L,P_L,type='o',col=col_prior,cex=3,lwd=5,pch=16)
points(L,P_E_d_L,type='o',col=col_like,cex=3,lwd=5,pch=16)
points(L,P_L_d_E,type='o',col=col_post,cex=3,lwd=5,pch=16)
legend(.65,1.2,yjust=1,xjust=0,
      legend=c('P(L): "dist. prior"','P(E|L): "func. verosimilitud"','P(L|E): "dist. posterior"'),
      pch=16,lwd=4,seg.len=2,pt.cex=1.75,cex=1.35,box.lty='blank',
      col=c(col_prior,col_like,col_post))
```
